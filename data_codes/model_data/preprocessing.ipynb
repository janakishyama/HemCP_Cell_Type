```python
import pandas as pd
```

#### Preprocessing Prediction Data set

```python
file_path = './predictions.csv'
df_predictions = pd.read_csv(file_path)
df_predictions.head()
```

```python
# Function to convert the file_name to the required date_time format
def convert_to_datetime(file_name):
    # Remove the '.csv' extension
    file_name = file_name.replace('.csv', '')
    # Replace '!' with '-' and ' ' with 'T' to make it a proper datetime format
    file_name = file_name.replace('!', '-').replace(' ', 'T')
    # Parse to datetime and format as 'YYYY-MM-DD HH:MM'
    date_time = pd.to_datetime(file_name, format='%Y-%m-%dT%H-%M-%S').strftime('%Y-%m-%d %H:%M')
    return date_time

# Apply the function to the file_name column
df_predictions['date_time'] = df_predictions['file_name'].apply(convert_to_datetime)


# Display the updated dataframe
df_predictions.head()
```

```python
df_sorted = df_predictions.sort_values(by='date_time')
df_sorted
```

```python
output_file_path = './sorted_predictions.csv'
df_sorted.to_csv(output_file_path, index=False)
```

### preprocessing sensor data

```python
new_file_path = './sensor_avg_merged.csv'
sensor_avg_merged = pd.read_csv(new_file_path)   
```

```python
# Convert the date_time column to datetime type
sensor_avg_merged['date_time'] = pd.to_datetime(sensor_avg_merged['date_time'])

# Set the date_time column as the index
sensor_avg_merged.set_index('date_time', inplace=True)

# Resample the data to one-minute intervals and interpolate
df_interpolated = sensor_avg_merged.resample('1T').interpolate(method='linear')

# Reset the index to make date_time a column again
df_interpolated.reset_index(inplace=True)


# Display the first few rows of the interpolated dataframe
df_interpolated.head()
```

```python
output_file_path = './interplolated_sensordata.csv'
df_interpolated.to_csv(output_file_path, index=False)
```

#### Join interpolated sensor data and sorted predictions

```python

# Load the provided files into dataframes
sensor_data_path = './interplolated_sensordata.csv'
predictions_path = './sorted_predictions.csv'

df_sensor = pd.read_csv(sensor_data_path)
df_predictions = pd.read_csv(predictions_path)

# Convert date_time columns to datetime type for merging
df_sensor['date_time'] = pd.to_datetime(df_sensor['date_time'])
df_predictions['date_time'] = pd.to_datetime(df_predictions['date_time'])

# Perform the merge operation using an outer join to retain all data
merged_df = pd.merge(df_sensor, df_predictions, on='date_time', how='outer')

# Sort by date_time to maintain chronological order
merged_df.sort_values(by='date_time', inplace=True)

# Display the first few rows of the merged dataframe
merged_df.head()
```

```python
output_file_path = './combined_modeldata.csv'
merged_df.to_csv(output_file_path, index=False)
```
