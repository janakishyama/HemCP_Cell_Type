```python
import pandas as pd
```

```python

# Load the data into a DataFrame
file_path = './combined_modeldata.csv'
df = pd.read_csv(file_path)
df.info()
```

```python
# The dataframe has 43 columns, and the columns to rename are located at the end of the dataframe
# Renaming columns 37 to 41
df = df.rename(columns={df.columns[-6]: 'filename', df.columns[-5]: 'Greenflag', df.columns[-4]: 'Greenround', df.columns[-3]: 'Others', df.columns[-2]: 'Redflag', df.columns[-1]: 'Redround'})

# Drop the column 'filename'
df = df.drop(columns=['filename'])

# Drop rows where all of 'Greenflag', 'Greenround', 'Others', and 'Redflag' are NaN or missing
df = df.dropna(subset=['Greenflag', 'Greenround', 'Others', 'Redflag', 'Redround'], how='all')


# Display the cleaned dataframe
df.head()
```

```python
# Fill missing values with 0
df_filled = df.fillna(0)
df_filled.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(input_features, target_variables, test_size=0.2, random_state=42)

# Initialize the base model
base_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Create the multi-output regressor
multi_output_model = MultiOutputRegressor(base_model)

# Train the model
multi_output_model.fit(X_train, y_train)

# Predict on the test set
y_pred = multi_output_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred, multioutput='raw_values')
r2 = r2_score(y_test, y_pred, multioutput='raw_values')

mse, r2
```
